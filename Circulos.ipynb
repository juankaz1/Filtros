# Circulos

```python
archivo = genfromtxt("circulos.dat",delimiter=" ")

x = archivo[:,0]
y = archivo[:,1]

figure(figsize=(15,15))
plot(x, y, "o", markersize = 1)
show()
```

```python
print "En la imagen hay " + str(len(x)) + " puntos graficados"
```

```python
ct = 1024 
h=6.0/ct
mat=zeros((ct,ct))
for n in range(len(x)):
    mat[ int(x[n]/h), int(y[n]/h)]+=1.0

fft_mat = fft2(mat)
freq = fftshift(fftfreq(ct,h)) 
fft_int = zeros(1 + ct/2)  
quant = zeros(1 + ct/2)  

for i in range(ct): 
    for j in range(ct):  
        k= int( ((j-ct/2)**2 + (i-ct/2)**2)**0.5 )
        if (k<=ct/2): 
            quant[k]+=1
            fft_int[k]+=abs(fft_mat[i,j])
            
Diametros=2*h*fft_int/quant

print len(Diametros)
```

```python
plot(Diametros)
```

```python
print "Aproximadamente el diametro de los circulos es: " + str(mean(Diametros))
```
