# Full moon

```python
img = plt.imread('full_moon.jpg')
dimy = img.shape[0]
dimx = img.shape[1]
print "The dimensions of the image are" ,dimy, "x", dimx
plt.imshow(img)
```

```python
intensity=zeros(img.shape[0:2])
for i in range(img.shape[0]):
    for j in range(img.shape[1]):
        intensity[i,j]=0.2126*img[i,j,0]+ 0.7152*img[i,j,1]+ 0.0722*img[i,j,2]
        
D1 = fft2(intensity)
```

#### En esta imagen se hará uso de la multiplicacion de bumps gauseanos de la forma $e^{-0.05t^2}$ y $ 1 - e^{-0.05t^2}$ para la filtracion

### ---------------------Pasa Bajas

```python
low_freq3 = fftshift(D1)
for i in range(D1.shape[0]):
    for j in range(D1.shape[1]):
        low_freq3[i,j]*=exp(-5E-2*((i-int(D1.shape[0])/2)**2 + (j-int(D1.shape[1])/2)**2))
        
im3_low = ifft2(fftshift(low_freq3,axes=(0,1)),axes=(0,1))
imshow(abs(im3_low), cmap=matplotlib.cm.Greys_r)
```

### -------------------Pasa Altas

```python
high_freq3 = fftshift(D1)
for i in range(D1.shape[0]):
    for j in range(D1.shape[1]):
        high_freq3[i,j]*=(1-exp(-5E-2*((i-int(D1.shape[0])/2)**2 + (j-int(D1.shape[1])/2)**2)))

im3_high= ifft2(fftshift(high_freq3,axes=(0,1)),axes=(0,1))
imshow(abs(im3_high),cmap=matplotlib.cm.Greys_r)
```

# colesterol-1

```python
img2 = plt.imread('colesterol-1.tif')
dimy, dimx = shape(img2)
print "The dimensions of the image are" ,dimx, "x", dimy
plt.imshow(img2,cmap=cm.Greys_r)
```

```python
D2= fft2(img2)
imshow(log(abs(fftshift(D2))))
colorbar()
```

#### Se realizó una grafica logaritmica para poder apreciar todo mejor. Ahora haremos lo mismo que se muestra en la referencia de Paul Burke, se tomarán cortes suaves del cilindro interior para frecuencias bajas y exterior para las altas.

### ---------------------Pasa Bajas

```python
wmin = 20
wmaj = 30
A = zeros((244,357))

JJ = fftshift(D2)

for i in range(244):
    for j in range(357):
        r = sqrt((j-357.0/2.0)**2+(i-122.0)**2)
        if(r < wmin):
            A[i,j] = JJ[i,j]
        elif(r < wmaj and r>=wmin):
            A[i,j]=JJ[i,j]*0.5*(1 + sin(pi*(0.5 - (r-wmin)/(wmaj-wmin))))
        else:
            A[i,j]=0.0
        
im2_low = ifft2(A)
imshow(abs(im2_low),  cmap=matplotlib.cm.Greys_r)   
```

### -------------------Pasa Altas

```python
wmin = 5
wmaj = 10
A = zeros((244,357))

JJ = fftshift(D2)

for i in range(244):
    for j in range(357):
        r = sqrt((j-357.0/2.0)**2+(i-122.0)**2)
        if(r > wmaj):
            A[i,j] = JJ[i,j]
        elif(r > wmin  and r<=wmaj):
            A[i,j]=JJ[i,j]*0.5*(1 - sin(pi*(0.5 + (r-wmin)/(wmaj-wmin))))
        else:
            A[i,j]=0.0
        
im2_high = ifft2(A)
imshow(abs(im2_high),  cmap=matplotlib.cm.Greys_r)
```

# 20_popc_cho009-1

```python
img3 = plt.imread('20_popc_cho009-1.tif')
dimy, dimx = shape(img3)
print "The dimensions of the image are" ,dimx, "x", dimy
plt.imshow(img3,cmap=cm.Greys_r)
```

#### Seguiremos con la primera tecnica de filtrado, puesto que los resultados graficos son mas satisfactorios

### ---------------------Pasa Bajas

```python
D3 = fft2(img3)
low_freq = fftshift(D3)
for i in range(D3.shape[0]):
    for j in range(D3.shape[1]):
        low_freq[i,j]*=exp(-5E-2*((i-int(D3.shape[0])/2)**2 + (j-int(D3.shape[1])/2)**2))
        
img3_low = ifft2(fftshift(low_freq))
imshow(abs(img3_low),  cmap=matplotlib.cm.Greys_r)
```

### ---------------------Pasa Altas

```python
D3 = fft2(img3)
high_freq = fftshift(D3)
for i in range(D3.shape[0]):
    for j in range(D3.shape[1]):
        high_freq[i,j]*= (1 - exp(-5E-2*((i-int(D3.shape[0])/2)**2 + (j-int(D3.shape[1])/2)**2)))
        
img3_high = ifft2(fftshift(high_freq))
imshow(abs(img3_high),  cmap=matplotlib.cm.Greys_r)
```
